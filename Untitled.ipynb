```python
import json
import os 
import numpy as np
from matplotlib import pyplot as plt

with open('/Users/glumli/Code/MT/app/out/serverlog/log/logFile_9_12_2021_10:16:15.log') as json_file:
    data = json.load(json_file)
    cohort = data["Hypertension"]    
    
options = {
    'bruteForce': {
        'color': 'red'
    }, 
    'resourceTypeIndex': {
        'color': 'violet'
    },
    'referenceIndex': {
        'color': 'blue'
    },
    'referenceIndexNew': {
        'color': 'cyan'
    },
    'optimum': {
        'color': 'green'
    },
}

def averageForPatient(patient, type = [''], metric='fetches'):
    fetches = []
    for key in patient:
        if(key.startswith(tuple(type))):
            res = patient[key][metric]
            if metric == 'resultSize':
                res = res / 1000000
            fetches.append(res)
    return np.mean(fetches)

def averageForAlgorithm(algorithmResults, type=['']):
    patientAverages = []
    for patientId in algorithmResults:
        patientAverages.append(averageForPatient(algorithmResults[patientId], tuple(type)))
    return np.mean(patientAverages)
        

def getResourceAmount(patientId, metric='fetches'):
    # for now we just assume that this testcase exists
    return cohort['bruteForce'][patientId]['val1'][metric]

def averageResourceAmount():
    resourceAverages = []
    for patientId in cohort['bruteForce']:
        resourceAverages.append(getResourceAmount(patientId))
    return np.mean(resourceAverages)

def scatterPlot(type=[''], skip=[], metric='fetches'):
    fig, ax = plt.subplots()
    for algorithmId in cohort:
        if(algorithmId in skip):
            continue
        algorithmResults = cohort[algorithmId]
        algorithmScatterY = []
        algorithmScatterX = []
        for patientId in algorithmResults:
            patientResults = algorithmResults[patientId]
            algorithmScatterY.append(averageForPatient(patientResults, type, metric))
            algorithmScatterX.append(getResourceAmount(patientId, 'fetches'))
        ax.scatter(
            algorithmScatterX,
            algorithmScatterY,
            color=options[algorithmId]['color'],
            label=algorithmId,
            s=5,
         #   marker='o',
        )
        m, b = np.polyfit(algorithmScatterX, algorithmScatterY, 1)
        ax.plot(algorithmScatterX, [m * x + b for x in algorithmScatterX], color=options[algorithmId]['color'], linewidth=0.5)
    
    # axe labels
    ax.set_xlabel("Number of Resources Available")
    if (metric=="fetches"):
        ax.set_ylabel("Number of Resources Fetched")
    else:
        ax.set_ylabel("Fetched Data in MB")
    ax.legend()
    ax.grid(True)
    plt.rcParams["figure.figsize"] = (6.4, 4.8)
    plt.savefig('out/papercharts/' + metric + '+'.join(type) + '.png')
    plt.plot()

def accelleration(algorithmResults, type=['']):
    Rretreived = averageForAlgorithm(algorithmResults, type)
    Rtotal = averageResourceAmount()
    return Rtotal / Rretreived

def precision(algorithmResults, type=['']):
    Rretreived = averageForAlgorithm(algorithmResults, type)
    Roptim = averageForAlgorithm(cohort['optimum'], type)
    return Roptim / Rretreived

def fallout(algorithmResults, type=['']):
    Rretreived = averageForAlgorithm(algorithmResults, type)
    Rtotal = averageResourceAmount()
    Roptim = averageForAlgorithm(cohort['optimum'], type)
    return (Rretreived - Roptim) / (Rtotal - Roptim) 

def efficiency(algorithmResults, type=['']):
    return precision(algorithmResults, type) * (1 - fallout(algorithmResults, type))

def printMetrics(type=['']):
    for algorithmId in cohort:
        print(algorithmId, ', '.join(type))
        print('accelleration: ', accelleration(cohort[algorithmId]))
        print('precision: ', precision(cohort[algorithmId]))
        print('fallout: ', fallout(cohort[algorithmId]))
        print('efficiency: ', efficiency(cohort[algorithmId]))
        print('')
    print('')
```
